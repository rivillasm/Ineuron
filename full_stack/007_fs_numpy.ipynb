```python
# -1 is a default number , so if your original dataset has 3x4 = 24 elements
# if you ask for 3,-1; -1 will produce such a value that 3 x suchavalue = 24
test4.reshape(3,-1)  # -1 is a default number
```

```python
test4.reshape(2,-1)  # -1 is a default number
# if you ask for 2,-1; -1 will produce such a value that 2 x suchavalue = 24, therefore 12
# at the end you will have an array 2x12 dimension
```
