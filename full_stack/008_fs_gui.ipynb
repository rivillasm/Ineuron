```python
# PACK()
import tkinter
root=tkinter.Tk()
root.geometry("200x600")
lbl1=tkinter.Label(root,text="Red Sun",bg="red", fg="white")
lbl2=tkinter.Label(root,text="Green grass",bg="green", fg="black")
lbl3=tkinter.Label(root,text="blue sky",bg="blue", fg="white")
lbl1.pack(fill=tkinter.X)  # files the entire X axis
lbl2.pack(fill=tkinter.X)
lbl3.pack(fill=tkinter.X)
root.mainloop()
```

```python
# PLACE()
import tkinter
root=tkinter.Tk()
root.geometry("200x600")
lbl1=tkinter.Label(root,text="Red Sun x:80 y:90",bg="red", fg="white")
lbl2=tkinter.Label(root,text="Green grass x:50 y:40",bg="green", fg="black")
lbl3=tkinter.Label(root,text="blue sky x:25 y:100",bg="blue", fg="white")
lbl1.place(x=80, y=90)  
lbl2.place(x=50, y=40)
lbl3.place(x=25, y=100)
root.mainloop()
```

```python
# PLACE
import tkinter
root=tkinter.Tk()
root.geometry("200x600")
lbl1=tkinter.Label(root,text="HEIGHT")
lbl2=tkinter.Label(root,text="WIDTH")
lbl1.place(x=5,y=5,width=100,height=50)
lbl2.place(x=5,y=60,width=100,height=50)
e1=tkinter.Entry(root)
e2=tkinter.Entry(root)
e1.place(x=120,y=5,width=200,height=50)
e2.place(x=120,y=60,width=200,height=50)
b1=Button(root,text="Zoom in")
b2=Button(root,text="Zoom in")
b1.place(x=5,y=120,width=100,height=50)
b2.place(x=120,y=120,width=100,height=50)

root.mainloop()
```

```python
from tkinter import *

def changecolor(e):
    ch=e.char
    if ch=="r":
        root.config(bg="red")
    elif ch=="b":
        root.config(bg="blue")
    elif ch=="g":
        root.config(bg="green")
        
root =Tk()
root.geometry("200x200")
root.bind("<Key>",changecolor)  # executes based on the clicked keyboard r,g,b
root.mainloop()
    
```

```python
from tkinter import *

def finish():
    root.destroy()
    
def show():
    fname=e1.get()
    lname=e2.get()
    fullname = fname+" "+lname
    l3.config(text=fullname)
    
root=tkinter.Tk()
root.geometry("200x200")
l1=Label(root,text="First Name:")
l2=Label(root,text="Last Name:")
l3=Label(root, width=20, anchor=W, font="Arial 10 bold")
e1=Entry(root)
e2=Entry(root)
b1=Button(root,text="Show",command=show)
b2=Button(root,text="Quit",command=finish)
l1.grid(row=0,column=0)
l2.grid(row=1,column=0)
e1.grid(row=0,column=1)
e2.grid(row=1,column=1)
l3.grid(row=2,columnspan=2, sticky=W)
b1.grid(row=3,column=0, sticky=W, pady=4)
b2.grid(row=3,column=1, sticky=W, pady=4)
root.mainloop()
```

```python
from tkinter import *

def finish():
    root.destroy()
    
def divide():
    a=float(e1.get())
    b=float(e2.get())    
    try:
        divide=a/b
        e3.insert(0,divide)
    except ZeroDivisionError:
        e3.config(bg="red",fg="black")
        e3.insert(0,"Division by Zero - Error")
        
    
def clear():
    e1.delete(0,END)
    e2.delete(0,END)
    e3.delete(0,END)
    e3.config(bg="white")
    
root=Tk()
root.geometry("400x200")
l1=Label(root,text="Number Division Program", font="Arial 18 bold")
l2=Label(root,text="First No:")
l3=Label(root,text="Second No:")
l4=Label(root,text="Result:")
e1=Entry(root)
e2=Entry(root)
e3=Entry(root)
b1=Button(root,text="Divide",command=divide)
b2=Button(root,text="Clear",command=clear)
b3=Button(root,text="Quit",command=finish)
l1.grid(row=0,column=0,columnspan=4)
l2.grid(row=1,column=0,sticky=E)
e1.grid(row=1,column=1)
l3.grid(row=2,column=0, sticky=E)
e2.grid(row=2,column=1)
l4.grid(row=3,column=0, sticky=E)
e3.grid(row=3,column=1)
b1.grid(row=4,column=0, pady=5, padx=5, sticky=E+W)
b2.grid(row=4,column=1, pady=5, padx=5, sticky=E+W)
b3.grid(row=4,column=2, pady=5, padx=5, sticky=E+W)
root.mainloop()

```

```python
from tkinter import *
from tkinter import messagebox

def finish():
    root.destroy()
    
def divide():
    try:
        e3.delete(0,END)
        a=int(e1.get())
        b=int(e2.get())    
        c=a/b
        e3.insert(0,str(c))
    except ZeroDivisionError:
        messagebox.showerror("Arithmetic Error!","Denominator should not be Zero!")
    except ValueError :
        messagebox.showerror("Conversion Error!","Only Digits Allowed!")
           
def clear():
    e1.delete(0,END)
    e2.delete(0,END)
    e3.delete(0,END)
    
root=Tk()
root.geometry("400x200")
l1=Label(root,text="Number Division Program", font="Arial 18 bold")
l2=Label(root,text="First No:")
l3=Label(root,text="Second No:")
l4=Label(root,text="Result:")
e1=Entry(root)
e2=Entry(root)
e3=Entry(root)
b1=Button(root,text="Divide",command=divide)
b2=Button(root,text="Clear",command=clear)
b3=Button(root,text="Quit",command=finish)
l1.grid(row=0,column=0,columnspan=4)
l2.grid(row=1,column=0,sticky=E)
e1.grid(row=1,column=1)
l3.grid(row=2,column=0, sticky=E)
e2.grid(row=2,column=1)
l4.grid(row=3,column=0, sticky=E)
e3.grid(row=3,column=1)
b1.grid(row=4,column=0, pady=5, padx=5, sticky=E+W)
b2.grid(row=4,column=1, pady=5, padx=5, sticky=E+W)
b3.grid(row=4,column=2, pady=5, padx=5, sticky=E+W)
root.mainloop()
```
