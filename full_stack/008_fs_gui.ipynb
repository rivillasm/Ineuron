```python
# PACK()
import tkinter
root=tkinter.Tk()
root.geometry("200x600")
lbl1=tkinter.Label(root,text="Red Sun",bg="red", fg="white")
lbl2=tkinter.Label(root,text="Green grass",bg="green", fg="black")
lbl3=tkinter.Label(root,text="blue sky",bg="blue", fg="white")
lbl1.pack(fill=tkinter.X)  # files the entire X axis
lbl2.pack(fill=tkinter.X)
lbl3.pack(fill=tkinter.X)
root.mainloop()
```

```python
# PLACE()
import tkinter
root=tkinter.Tk()
root.geometry("200x600")
lbl1=tkinter.Label(root,text="Red Sun x:80 y:90",bg="red", fg="white")
lbl2=tkinter.Label(root,text="Green grass x:50 y:40",bg="green", fg="black")
lbl3=tkinter.Label(root,text="blue sky x:25 y:100",bg="blue", fg="white")
lbl1.place(x=80, y=90)  
lbl2.place(x=50, y=40)
lbl3.place(x=25, y=100)
root.mainloop()
```

```python
from tkinter import *

def changecolor(e):
    ch=e.char
    if ch=="r":
        root.config(bg="red")
    elif ch=="b":
        root.config(bg="blue")
    elif ch=="g":
        root.config(bg="green")
        
root =Tk()
root.geometry("200x200")
root.bind("<Key>",changecolor)  # executes based on the clicked keyboard r,g,b
root.mainloop()
    
```

```python
from tkinter import *

def finish():
    root.destroy()
    
def show():
    fname=e1.get()
    lname=e2.get()
    fullname = fname+" "+lname
    l3.config(text=fullname)
    
root=tkinter.Tk()
root.geometry("200x200")
l1=Label(root,text="First Name:")
l2=Label(root,text="Last Name:")
l3=Label(root, width=20, anchor=W, font="Arial 10 bold")
e1=Entry(root)
e2=Entry(root)
b1=Button(root,text="Show",command=show)
b2=Button(root,text="Quit",command=finish)
l1.grid(row=0,column=0)
l2.grid(row=1,column=0)
e1.grid(row=0,column=1)
e2.grid(row=1,column=1)
l3.grid(row=2,columnspan=2, sticky=W)
b1.grid(row=3,column=0, sticky=W, pady=4)
b2.grid(row=3,column=1, sticky=W, pady=4)
root.mainloop()
```

```python
from tkinter import *

def finish():
    root.destroy()
    
def divide():
    a=float(e1.get())
    b=float(e2.get())    
    try:
        divide=a/b
        e3.insert(0,divide)
    except ZeroDivisionError:
        e3.config(bg="red",fg="black")
        e3.insert(0,"Division by Zero - Error")
        
    
def clear():
    e1.delete(0,END)
    e2.delete(0,END)
    e3.delete(0,END)
    e3.config(bg="white")
    
root=Tk()
root.geometry("400x200")
l1=Label(root,text="Number Division Program", font="Arial 18 bold")
l2=Label(root,text="First No:")
l3=Label(root,text="Second No:")
l4=Label(root,text="Result:")
e1=Entry(root)
e2=Entry(root)
e3=Entry(root)
b1=Button(root,text="Divide",command=divide)
b2=Button(root,text="Clear",command=clear)
b3=Button(root,text="Quit",command=finish)
l1.grid(row=0,column=0,columnspan=4)
l2.grid(row=1,column=0,sticky=E)
e1.grid(row=1,column=1)
l3.grid(row=2,column=0, sticky=E)
e2.grid(row=2,column=1)
l4.grid(row=3,column=0, sticky=E)
e3.grid(row=3,column=1)
b1.grid(row=4,column=0, pady=5, padx=5, sticky=E+W)
b2.grid(row=4,column=1, pady=5, padx=5, sticky=E+W)
b3.grid(row=4,column=2, pady=5, padx=5, sticky=E+W)
root.mainloop()

```

```python
from tkinter import *
from tkinter import messagebox

def finish():
    answer = messagebox.askyesno("Are you sure you want to quit?")
    if answer == True:
        root.destroy()
    
def divide():
    try:
        e3.delete(0,END)
        a=int(e1.get())
        b=int(e2.get())    
        c=a/b
        e3.insert(0,str(c))    # "0" is an index
    except ZeroDivisionError:
        messagebox.showerror("Arithmetic Error!","Denominator should not be Zero!")
    except ValueError :
        messagebox.showerror("Conversion Error!","Only Digits Allowed!")
           
def clear():
    e1.delete(0,END)
    e2.delete(0,END)
    e3.delete(0,END)
    e1.focus()   # cursor goes back to e1
    root.config(bg=oldcolor)  #
    
root=Tk()
root.geometry("400x200")
l1=Label(root,text="Number Division Program", font="Arial 18 bold")
l2=Label(root,text="First No:")
l3=Label(root,text="Second No:")
l4=Label(root,text="Result:")
e1=Entry(root)
e2=Entry(root)
e3=Entry(root)
b1=Button(root,text="Divide",command=divide)
b2=Button(root,text="Clear",command=clear)
b3=Button(root,text="Quit",command=finish)
l1.grid(row=0,column=0,columnspan=4)
l2.grid(row=1,column=0,sticky=E)
e1.grid(row=1,column=1)
l3.grid(row=2,column=0, sticky=E)
e2.grid(row=2,column=1)
l4.grid(row=3,column=0, sticky=E)
e3.grid(row=3,column=1)
b1.grid(row=4,column=0, pady=5, padx=5, sticky=E+W)
b2.grid(row=4,column=1, pady=5, padx=5, sticky=E+W)
b3.grid(row=4,column=2, pady=5, padx=5, sticky=E+W)
root.mainloop()
```

```python
from tkinter import *
from tkinter import filedialog, messagebox

def showopen():
    file_names=filedialog.askopenfilenames(title="select your fav song", filetypes=[("mp3 files","A*.mp3")])
    str=""
    if type(file_names) is tuple:
        for file in file_names:
            str+=file+"\n"
        lbl.configure(text=str)
    else:
        messagebox.showinfo("Your selections","you did not select any file")
        
root=Tk()
root.geometry("200x200")
btn=Button(root,text="open file",command=showopen)
btn.pack()
root.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox

def showgender():
    value=x.get()
    if value == 1:
        messagebox.showinfo("Gender","You selected Male!")
    else:
        messagebox.showinfo("Gender","You selected Female!")
    

root=Tk()
root.geometry("400x200")
x=IntVar()
l1=Label(root,text="Select your Gender")
rb1=Radiobutton(root,text="Male",command=showgender,variable=x,value=1)
rb2=Radiobutton(root,text="Female",command=showgender,variable=x,value=2)
l1.grid(row=0,column=0)
rb1.grid(row=1,column=1, sticky=W)
rb2.grid(row=2,column=1, sticky=W)

root.mainloop()
```

```python
from tkinter import *
from tkinter import simpledialog, messagebox

def showitem():
    pos=simpledialog.askinteger("input","enter item no",minvalue=1,maxvalue=lb1.size())
    if type(pos) is int:
        item=lb1.get(pos-1)
        l1.configure(text="you selected: "+item)
    else:
        messagebox.showinfo("Cancel pressed!","you didnt imput any value") 
                                

root=Tk()
root.geometry("400x300")
l1=Label(root,text="your selection will appear here")
b1=Button(root,text="show item",command=showitem)
lb1=Listbox(root)
sports=["cricket","football","hockey","basketball","tennis"]
for x in sports:
    lb1.insert(END,x)
    
lb1.grid(row=0,column=0,sticky=W)
b1.grid(row=1,column=0,sticky=W)
l1.grid(row=2,column=0,sticky=W)
root.mainloop()
```

```python
from tkinter import *

def setredcolor(val):
    red=hex(int(val))
    red=red[2:]
    if len(red)==1:
        red="0"+red
    color="#"+red+"0000"
    root.configure(bg=color)
    
root=Tk()
root.geometry("200x200")
sc=Scale(root,from_=0,to=255,orient=HORIZONTAL,command=setredcolor)
sc.pack()
root.configure(bg="#000")
root.mainloop()
         
```

```python
from tkinter import *
from tkinter import messagebox,simpledialog

class mygui:
    def __init__(self,root):
        self.root=root
        self.root.geometry("420x200+100+20")
        self.l1=Label(root,text="Number Division Program",font="Arial 18 bold")
        self.l2=Label(root,text="First No:")
        self.l3=Label(root,text="Second No:")
        self.l4=Label(root,text="Result:")
        self.e1=Entry(root)
        self.e2=Entry(root)
        self.e3=Entry(root)
        self.e4=Entry(root)
        self.b1=button(root,text="Divide",command=self.divide)
        self.b2=button(root,text="Clear",command=self.clear)
        self.b3=button(root,text="Quit",command=self.finish)
        self.l1.grid(row=0,colum=0,columnspan=4)
        self.l2.grid(row=1,colum=0,sticky=E)
        self.e1.grid(row=1,colum=1)
        self.l3.grid(row=2,colum=0,sticky=E)
        self.e2.grid(row=2,colum=1)
        self.l4.grid(row=3,colum=0,sticky=E)
        self.e3.grid(row=3,colum=1)
        self.l3.grid(row=2,colum=0,sticky=E)
        self.b1.grid(row=4,colum=0, pady=5, padx=5, sticky=E+W)
        self.b2.grid(row=4,colum=1, pady=5, padx=5, sticky=E+W)
        self.b3.grid(row=4,colum=2, pady=5, padx=5, sticky=E+W)
        #self.root.mainloop()
        
    def divide(self):
        try:
            self.e3.delete(0,END)
            self.e3.config(fg="red")
            a=int(self.e1.get())
            b=int(self.e2.get())
            c=a/b
            self.e3.insert(0,str(c))
            self.e3.config(fg="green")
        except ZeroDivisionError:
            messagebox.showerror("Arithmetic Error!","Denominator should not be 0!")
        except ValueError:
            messagebox.showerror("Conversion Error!","Only digits allowed!")
            
    def clear():
        self.e1.delete(0,END)
        self.e2.delete(0,END)
        self.e3.delete(0,END)
        self.e1.focus()   # cursor goes back to e1
        
    def finish():
        answer = messagebox.askyesno("Quitting!","Are you sure you want to quit?")
        if answer == True:
            name=simpledialog.askstring("Input","What is your name?")
        if name is None or name=="":
            name=" UserJee!"
        messgebox.showinfo("Thank you","Have a good day "+name)
        self.root.destroy()
   
```
