```python
from datetime import *
from tkinter import *

global count
count=0

def showmsg():
    global count 
    #count=0
    count+=1
    l1=tkinter.Label(root,text=datetime.now().date().strftime("%d-%B-%Y"))
    l2=tkinter.Label(root,text=count)
    l1.pack()
    l2.pack()
    
    
root = tkinter.Tk()
root.geometry("200x200")
b1=Button(root,text="click me", command=showmsg)
b1.pack(side="bottom")
root.mainloop()
```

* pack() -

```python
# PACK()
import tkinter
root=tkinter.Tk()
root.geometry("200x600")
lbl1=tkinter.Label(root,text="Red Sun",bg="red", fg="white")
lbl2=tkinter.Label(root,text="Green grass",bg="green", fg="black")
lbl3=tkinter.Label(root,text="blue sky",bg="blue", fg="white")
lbl1.pack(fill=tkinter.X)  # files the entire X axis
lbl2.pack(fill=tkinter.X)
lbl3.pack(fill=tkinter.X)
root.mainloop()
```

```python
# PLACE()
import tkinter
root=tkinter.Tk()
root.geometry("200x600")
lbl1=tkinter.Label(root,text="Red Sun x:80 y:90",bg="red", fg="white")
lbl2=tkinter.Label(root,text="Green grass x:50 y:40",bg="green", fg="black")
lbl3=tkinter.Label(root,text="blue sky x:25 y:100",bg="blue", fg="white")
lbl1.place(x=80, y=90)  
lbl2.place(x=50, y=40)
lbl3.place(x=25, y=100)
root.mainloop()
```

```python
# PLACE
import tkinter
root=tkinter.Tk()
root.geometry("200x600")
lbl1=tkinter.Label(root,text="HEIGHT")
lbl2=tkinter.Label(root,text="WIDTH")
lbl1.place(x=5,y=5,width=100,height=50)
lbl2.place(x=5,y=60,width=100,height=50)
e1=tkinter.Entry(root)
e2=tkinter.Entry(root)
e1.place(x=120,y=5,width=200,height=50)
e2.place(x=120,y=60,width=200,height=50)
b1=Button(root,text="Zoom in")
b2=Button(root,text="Zoom in")
b1.place(x=5,y=120,width=100,height=50)
b2.place(x=120,y=120,width=100,height=50)

root.mainloop()
```

```python
# GRID
import tkinter
root=tkinter.Tk()
root.geometry("200x600")
lbl1=tkinter.Label(root,text="NAME:")
lbl2=tkinter.Label(root,text="PASW:")
e1=tkinter.Entry(root)
e2=tkinter.Entry(root)
b1=Button(root,text="LOGIN")
b2=Button(root,text="QUIT")

lbl1.grid(row=0,column=0,sticky=tkinter.W)
lbl2.grid(row=1,column=0)
e1.grid(row=0,column=1)
e2.grid(row=1,column=1)
b1.grid(row=2,column=0)
b2.grid(row=2,column=1)

root.mainloop()
```

```python
# GRID
import tkinter
root=tkinter.Tk()
root.geometry("200x200")
lbl1=tkinter.Label(root,text="NAME:")
lbl2=tkinter.Label(root,text="PASW:")
e1=tkinter.Entry(root)
e2=tkinter.Entry(root)
b1=Button(root,text="LOGIN")
b2=Button(root,text="QUIT")
c1=tkinter.Checkbutton(root,text="keep me logged in")

lbl1.grid(row=0,column=0,sticky=tkinter.E)
e1.grid(row=0,column=1,columnspan=2)
lbl2.grid(row=1,column=0)
e2.grid(row=1,column=1,columnspan=2)
c1.grid(row=2,column=0,columnspan=2)
b1.grid(row=3,column=1,sticky=tkinter.W+tkinter.E)
b2.grid(row=3,column=2,sticky=tkinter.W+tkinter.E)

root.mainloop()
```
