# MAPPING

```python
l=[1,2,3,4,5,6]
l
```

```python
ll=[]
for i in l:        # for loop
    ll.append(i+5)
ll
```

```python
[i+5 for i in l]    # comprehensive operation
```

```python
def test(a):
    return a+5

map(test,l)  # outputs an iterative
```

```python
list(map(test,l))
```

```python
l1=["234","235","236","237"]
```

```python
[int(i) for i in l1]
```

```python
def test1(a):
    return int(a)

list(map(test1,l1))
```

```python
list(map(lambda x : int(x), l1))    # lambda function
```

```python
list(map(lambda x : x+5, l))    # lambda function
```

# FILTER

```python
l2 = [4,5,6,7,78,8,78]
```

```python
[i for i in l2 if i%2==0]     # only even numbers filtered
```

```python
def test4(a):
    if a%2==0:
        return a

list(filter(test4,l2))
```

* Map()
The basic function of map() is to manipulate iterables. Map executes all the conditions of a function on the items in the iterable.
*Filter()
As the name suggests, it is used to filter the iterables as per the conditions. Filter filters the original iterable and passes the items that returns True for the function provided to filter. 



* Map takes all objects in a list and allows you to apply a function to it whereas Filter takes all objects in a list and runs that through a function to create a new list with all objects that return True in that function

```python
list(filter(lambda var:var%2 == 0, l2))
```

```python
list(filter(lambda var : var+5, l2))    # here var+5 will always TRUE for each element, 
                                        # therefore all get passed
```

# REDUCE

```python
from functools import reduce
```

```python
l = [0,4,5,6,7,3,8]
sum(l)
```

```python
def test5(a,b):
    return a+b
reduce(test5,l)
```

* reduce ONLY takes two items at the time, you may try with 3 but not neceseraly is going to work when the amount of elements (data input) in the list do not support that splitting ...



```python
reduce(lambda x,y : x+y, l)      # with lambda
```
