```python
class bonouscalculator:
    def __init__(self,empid,emprating):
        self.empid=empid
        self.emprating=emprating
        self.__bonusforratingA="70%"  # private variable
        self.__bonusforratingB="60%"
        self.__bonusforratingC="40%"
        
    def bonuscalculator(self):
        if self.emprating=="A":
            bonus = self.__bonusforratingA   # i can access the private variable within the class
            return bonus
        elif self.emprating=="B":
            bonus = self.__bonusforratingB
            return bonus
        else:
            bonus = self.__bonusforratingC
            return bonus
        
        
```

```python
emp1 = bonouscalculator(101,"A")
emp2 = bonouscalculator(102
                    ,"A")
emp3 = bonouscalculator(103,"A")
```
