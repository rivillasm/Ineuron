###  Class is a blue print, entity is an object that exists.

Classes provide a means of bundling data and functionality together. Creating a new class creates a new type of object, allowing new instances of that type to be made. Each class instance can have attributes attached to it for maintaining its state. Class instances can also have methods (defined by its class) for modifying its state.

```python
class car:                  #  creates a class
    # if starts and finishes with __ this is an inbuilt function created by someone 
    # required to create objects
    def __init__(self, brand_name, fuel_type, body_type):  # constructor __init__ helps my class to get an input
        self.brand_name=brand_name
        self.fuel_type=fuel_type
        self.body_type=body_type
        
    def desc_car(self):          # adding self allows this function to be used as a mode in the class
        print(self.brand_name, self.fuel_type, self.body_type)
                            
```

```python
innova = car("toyota", "gas", "cadilack")   # creates a variable/instance of a car type
```

```python
class list_parser:
    def __init__(self, l):
        self.l=l
        
    def parcer(self):            # self pointer required so we can use the values of the class
        if type(self.l)==list:   # needs the pointer self to access the class values in other functions
            for i in self.l:
                print(i)
                
    def reverser_list(self):     # # self pointer required so we can use the values of the class
        if type(self.l)==list:   # needs the pointer self to access the class values in other functions
            return self.l[::-1]
```

```python
c=list_parser([2,3,4,5,6])
c.l
```

```python
c.parcer()
```

```python
c.reverser_list()
```

```python
class dict_parser:
    
    def __init__(self,d):
        self.d=d
        
    def _keys(self):
        if self._nodict():
            return self.d.keys()
        
    def _values(self):
        if self._nodict():
            return self.d.values()
        
    def _nodict(self):
        if type(self.d)!=dict:
            raise Exception(self.d, " this is not a dictionary ")
        return 1
            
    def _userinput(self):
        self.d =eval( input("input a dictionary")) # evaluates the input
        if self._nodict():
            print(self.d, type(self.d))
            print(self.d.keys())
            print(self.d.values())
            
    def _insert(self):
        if self._nodict():
            k,v = input("add a new key/value pair:").split()
            self.d[k]=v
            print(self.d)
            
    def _insert1(self, **kwargs):   # input several key/value pairs
        for k,v in kwargs.items():
            self.d[k],v
            print(self.d)
            
            
```
