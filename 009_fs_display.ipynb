```python
df= pd.read_csv('https://raw.githubusercontent.com/venky14/Machine-Learning-with-Iris-Dataset/master/Iris.csv')
df
```

```python
df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]].plot(figsize=(20,10))
```

```python
df.loc[5:11,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]].plot(kind="bar",figsize=(20,10))
```

```python
df.loc[5:11,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]].plot(kind="barh",figsize=(20,10))
```

```python
df.loc[:,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]].plot(kind="hist",figsize=(20,10))
```

```python
df.loc[:,["SepalLengthCm"]].plot(kind="hist",orientation='horizontal',figsize=(20,10))
```

```python
df.hist(figsize=(10,10),color="g",alpha=.4)
```

```python
df.loc[:,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]].plot(figsize=(20,15),kind="box")
```

```python
df.loc[:,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]].plot(figsize=(20,15),color={"boxes":"g","whiskers":"r"},kind="box")
```

```python
df.loc[:,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"]].plot(kind="area",figsize=(15,15))
```

```python
df.plot.scatter(x="SepalLengthCm",y="PetalLengthCm",figsize=(10,10))
```

```python
df.plot.scatter(x="SepalLengthCm",y="PetalLengthCm",c="SepalWidthCm",figsize=(10,10),s=80)
```

```python
df.plot.hexbin(x="SepalLengthCm",y="PetalLengthCm",figsize=(10,10),gridsize=10)
```

```python
df.drop(columns=["Species","Id"]).iloc[0].plot(kind="pie",figsize=(10,10))
```
