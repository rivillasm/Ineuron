```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import linear_model
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({"area":[2600, 3000, 3200, 3600, 4000], "price":[550000, 565000,610000,680000,725000]})
df
```

```python
%matplotlib inline
fig=plt.figure(figsize=(8,6))
plt.scatter(df.area, df.price, color="r", marker="o")
plt.xlabel("area [Ft**2]")
plt.ylabel("price [USD]")
```

```python
df[["area"]].shape
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)    # expects 2D array
```

```python
 reg.predict([[3300]])      # expects 2D array
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# y = intercep + slope * Xi
reg.intercept_ + reg.coef_ * 3300
```

```python
df_test = pd.DataFrame({"area":[1000, 1500, 2300, 3540, 4120, 4560, 5490, 3460, 4750, 2300, 9000, 8600, 7100]})
df_test
```

```python
df_predicted = reg.predict(df_test)
df_predicted = reg.predict(df_test[["area"]])
```

```python
df_test["predicted"] = df_predicted.copy()
df_test
```

```python
%matplotlib inline
fig=plt.figure(figsize=(8,6))
plt.scatter(df.area, df.price, color="r", marker="o")
plt.plot(df.area, reg.predict(df[["area"]]), color="b")
plt.xlabel("area [Ft**2]",fontsize=20)
plt.ylabel("price [USD]",fontsize=20)
```
