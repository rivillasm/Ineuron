```python
# merging two data frames
df1=pd.DataFrame({'a':['a0','a1','a2','a3'],
                  'b':['b0','b1','b2','b3'],
                  'c':['c0','c1','c2','c3'],
                  'd':['d0','d1','d2','d3']}, index=[0,1,2,3])
df1
```

```python
df2=pd.DataFrame({'a':['a4','a5','a6','a7'],
                  'b':['b4','b5','b6','b7'],
                  'c':['c4','c5','c6','c7'],
                  'd':['d4','d5','d6','d7']}, index=[4,5,6,7])
df2
```

```python
df3=pd.DataFrame({'a':['a8','a9','a10','a11'],
                  'b':['b8','b9','b10','b11'],
                  'c':['c8','c9','c10','c11'],
                  'd':['d8','d9','d10','d11']}, index=[8,9,10,11])
df3
```

```python
# concatenation
df_cat1=pd.concat([df1,df2,df3], axis=0)  # row after row, indexes names are kept
df_cat1
```

```python
# concatenation
df_cat1=pd.concat([df1,df2,df3], axis=1)  # col after col, indexes names are kept
df_cat1  # since we have diff indexes
```

```python
# merging by a common key

left = pd.DataFrame({'key':['k0','k8','k2','k3'],
                     'a':['a0','a1','a2','a3'],
                     'b':['b0','b1','b2','b3']})

left
```

```python
right = pd.DataFrame({'key':['k0','k1','k2','k3'],
                     'c':['c0','c1','c2','c3'],
                     'd':['d0','d1','d2','d3']})

right
```

```python
merge1 = pd.merge(left,right,how="inner",on='key')   # inner - what there is in common.
merge1
```

```python
merge1 = pd.merge(left,right,how="outer",on='key')   # outer - ALL together. order of input datasets matter
merge1
```

```python
left1 = pd.DataFrame({'key1':['k0','k8','k2','k3'],
                     'a':['a0','a1','a2','a3'],
                     'b':['b0','b1','b2','b3']})

left1
```

```python
right1 = pd.DataFrame({'key2':['k0','k1','k2','k3'],
                     'c':['c0','c1','c2','c3'],
                     'd':['d0','d1','d2','d3']})

right1
```

```python
# when the keys have different names
merge2 = pd.merge(left1,right1,how="outer",left_on='key1',right_on='key2')   # outer - ALL together. order of input datasets matter
merge2
```

```python
# left merges only the items in dataset1
merge3 = pd.merge(left1,right1,how="left",left_on='key1',right_on='key2')   # outer - ALL together. order of input datasets matter
merge3
```

```python
# right merges only the items in dataset1
merge4 = pd.merge(left1,right1,how="right",left_on='key1',right_on='key2')   # outer - ALL together. order of input datasets matter
merge4
```
