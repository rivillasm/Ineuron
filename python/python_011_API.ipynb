# API - application programming interface
* open pycharm, create a new project, give it a name

* on the created project, right click create new/file and give it a name

* click on the terminal tab at the left bottom
* install flask


Go to POSTMAN
and click on new/http request


* in the pycharm file, right click and run the file
* in the terminal see the http address
* 127.0.0.1 is the local machine and the 5000 is the flask application running at port




* in POSTMAN
    * select POST (we are providing values)
    * add the url and the port numbers (127.0.0.1:5000)
    * add the function name (math_route)
    * go to the body tab, and select raw/JSON
    * add the expected variables withthe corresponding values
    * click on send
