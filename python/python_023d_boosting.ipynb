# AdaBoos

* update weights
* use decision trees (only one dept - STUMPS) for training models 
* for each featrue we create a STUMP
* use ENTROPY/GINI to select the first decision tree
* once a stump selected, and once the classification of the input data is done by the decision tree, we need to find the TOTAL ERROR
* TOTAL ERROR (TE) - sum of all the sample weights of the erroneously classified data points (rows)
* find  the PERFROMANCE OF THE STUMP =  (log (1-TE/TE) )/2

* the whole point is to increase the weights of the wrongly classified data points and decrease the weights of the correctly classified data points
* updat the weights for the incorrect and correct classified data points

* update the weights for the correctly classified points


* sum all the new updated weights and normalize to one, so that the sume of all of them is equal to 1


* once the new weights are normalized is time to create a new dataset. to do that we create buckets of values (binnings)

* then the algorithm will pick and chose values from this input dataset based on the new normalized weight.
* this will create a complete new dataset where the data points with buckes with bigger values (wrongly classified previously) have more chance to be chosen, this way the algorithm ensures to pass the wrongly classified points into the new dataset for further analyze

* then the process repeaats itself, once the new dataset is selected, new STUMPS are created for each feature and by using ENTROPHY/GINI we will select what will be the first feature for the first STUMP, and so on

* once this process is done several times, the final solution will be given by ensembling these all weak learners and creating a strong learner from them in a sequential manner



# GRADIENT BOOSTING

* compute the BASE MODEL which will provide one output (average of all the target feature) (Yhat COLUMN IN THIS CASE)
* Compute RESIDUAL (PSEUDO RESIDUALS) [actgual value - predicted value] ( R1 COLUMN IN THIS CASE)
* construct a new DECISION TREE where the input will be the same imput features but the OUTPUT feature is NOW THE RESIDUALS .
* once the model is trained ofcoure the predictions will be RESIDUALS which will be added to the BASE MODEL output
* To avoid overfitting a LEARNING RATE WILL BE USED [Yhat + LEARNING_RATE * R2]

* Then a new decision tree is trained based on the output values R2 and the process keeps going

* so we are adding decision trees sequentially and finding residual in each step
* the goal is to add all of those results together considering the corresponding LEARNING RATES to avoid OVERFITTING
* ONCE the model is ready a new input data point will be run through all of those decision trees to get the final predicted answer
