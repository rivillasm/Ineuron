```python
class car:
    def __init__(self, year, make, model, engine_type, tyre_type):
        self.year1=year       # stores the data in the class
        self.make1=make
        self.model1=model
        self.engine_type1=engine_type
        self.tyre_type1=tyre_type
    def __str__(self):
        return "this is my car class for ford"
    def engineer(self):
        return self.year1
```

```python
ford = car(2020,"2342","edge","4cl","micheline")
```

```python
ford.make1
```

```python
print(ford)
```

```python
ford.engineer()
```

## Encapsulation

```python
class test2:
    def __init__(self, a,b,c):
        self._a=a
        self.__b=b
        self.c=c
```

```python
r = test2("mario","alejandro","rivillas")
```

* r.c - public
* r._a - protected
* r.__b - private



```python
# how to access a protected variable in a class
r._a
```

```python
# how to access a private variable in a class
r._test2__b
```

## inheritance

```python
class test3:
    def __init__(self, a,b,c,d):
        self.a=a
        self.b=b
        self.c=c
        self.d=d
    
    def test_in(self):
        return self.a
```

```python
class test4(test3):  # test4 inherits what is in test3
    def __init__(self, a,b,c,d):
        test3.__init__(self,a,b,c,d)    #allows seeing a,b,c,d in the test4 class
```

```python
r = test4(1,2,3,4)
```

```python
r.a
```
