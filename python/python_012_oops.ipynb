## inheritance of several classes

```python
class test:
    def __init__(self, a,b):
        self.a=a
        self.b=b
        
class test1:
    def __init__(self, c,d):
        self.c=c
        self.d=d
```

```python
class child(test,test1):                # sets inheritance from classes test,test1
        def __init__(self, *args):
            test.__init__(self,*args)   # calls class test
            test1.__init__(self,*args)  # calls class test1
            
        def fun_child(self):
            print(self.a, self.c)
```

```python
obj = child(7,2)
```

```python
obj.fun_child()
```

```python
class parent1:
    def __init__(self, a,b):
        self.a=a
        self.b=b
    
    def age(self):
        return self.a + self.b

class parent2(parent1):
    def __init__(self, c, *args):
        parent1.__init__(self,*args)   # calls class parent1
        self.c=c
        
class child1(parent2):
    def __init__(self,*args):
        parent2.__init__(self,*args)  # calls class parent2
```

```python
r = child1(3,5,7)
```

## Method overwritting

```python
class person:
    def __init__(self, name,surname,birth):
        self.name=name
        self.surname=surname
        self.birth=birth
    
    def age(self, current_year):
        return current_year - self.birth
    
    def __str__(self):
        return self.name
```

```python
per_object=person("mario","rivillas",1969)
```

```python
print(per_object)   # invoques the str function in the person class
```

```python
class student(person):
    def __init__(self, student_id, *args):
        super(student,self).__init__(*args)
        self.student_id=student_id
        
    def __str__(self):
        return (self.name + str(self.student_id))
    
```

```python
stud_obj = student(1500,"mario","rivillas",1969)
```

```python
print(stud_obj)
```

```python
# invoquing the __str__ function from the person class
# this is call method overwritting

class student(person):
    def __init__(self, student_id, *args):
        super(student,self).__init__(*args)
        self.student_id=student_id
        
    def __str__(self):
        return (super(student,self).__str__() + str(self.student_id))  # invoquint str
    
```

* Here we are invoquing the str from the parent class (person) plus the str from the current class (student). in other words we are overwritting it

```python
stud_obj = student(1500,"mario1","rivillas1",1969)
```

```python
print(stud_obj)
```

## encapsulation

```python
class tyre:
    def __init__(self, branch, belted_basis, opt_pressure):
        self.branch=branch
        self.belted_basis=belted_basis
        self.opt_pressure=opt_pressure
        
    def __str__(self):
        return self.branch + self.belted_basis + self.opt_pressure
```

```python
class engine:
    def __init__(self, fuel_type, noise_level):
        self.fuel_type=fuel_type
        self.noise_level=noise_level
    
    def __str__(self):
        return self.fuel_type + self.noise_level
    
class body:
    def __init__(self, size):
        self.size=size
    def __str__(self):
        return self.size
```

```python
class car:
    def __init__(self, a, b, c):
        self.a=a
        self.b=b
        self.c=c
        
    def __str__(self):
        return str(self.a)+str(self.b)+str(self.c)
```

```python
t = tyre(" branch1 ", " belted1 ", " 32psi ")
e = engine(" premium "," low_noise ")
b = body(" medium_size ")
```

```python
print(t)
```

```python
print(b)
```

```python
c=car("a","b","c")
```

```python
print(c)   # prints the str info for the corresponding class
```

```python
c1 = car(t,e,b)   # passing objects into the class car object
                  # so we are calling the str from the other classes from the car class
                  # this is called encapsulation
```

```python
print(c1)
```

# overloading

```python
class multiplynum():
    def __init__(self,a):
        self.a=a
        
    def __mul__(self, other):
        return self.a+other.a  # overloaded / overwritten the * for + for an existing mult function
```

```python
a1=multiplynum(4)
a2=multiplynum(5)
```

```python
# answers is 4+5=9 instead of 4*5=20 since we overloaded the 
a1*a2
```
