```python
import requests
from bs4 import BeautifulSoup as bs

# urllib is a Python module that can be used for opening URLs. 
# It defines functions and classes to help in URL actions.
from urllib.request import urlopen as uReq
```

## 1 - creates url

```python

searchString="mobile"   # select the item you want to search for

flipkart_url = "https://www.flipkart.com/search?q=" + searchString   # access an url for specific string
flipkart_url
```

## 2 - request/connect to the url

```python
uClient = uReq(flipkart_url)   # url request
uClient
```

## 3- read the info

```python
flipkartPage = uClient.read()
```

```python
flipkartPage
```

## 4- extract info

```python
flipkart_html = bs(flipkartPage, "html.parser")   # beautiful soup
flipkart_html
```

extracting the info from this class id obtained from the website inspector



```python
bigboxes = flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"})
bigboxes
```
